```python
import numpy as np
import pandas as pd
import requests
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

resp = requests.get('https://data.covid19.go.id/public/api/update.json', verify=False).json()
data = pd.DataFrame(resp['update']['harian'])
data_tidy = data[['key_as_string', 'jumlah_meninggal']]

plt.clf()
data_tidy['key_as_string'] = pd.to_datetime(data['key_as_string'], unit='ns').dt.date
data_tidy['jumlah_meninggal'] = pd.json_normalize(data.jumlah_meninggal)

fig, ax = plt.subplots(figsize=(10,5))
ax.plot('key_as_string', 'jumlah_meninggal', data=data_tidy, lw=2)

ax.set_title('Deaths of Indonesian Covid-19 Disease Statistics',fontsize=22)
ax.set_xlabel('')
ax.set_ylabel('Total Deaths')
ax.text(1, -0.5, 'Source: covid.19.go.id', color='blue',ha='right', transform=ax.transAxes)
ax.set_xticklabels(ax.get_xticks(), rotation=90)
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

plt.grid()
plt.tight_layout()
plt.show()
```
